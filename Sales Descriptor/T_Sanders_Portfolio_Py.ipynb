```python
vg_sales.loc[vg_sales.Platform != np.nan].groupby(['Platform']).agg({'Global_Sales':'sum'}).sort_values('Global_Sales', ascending = False)
```

```python
PS2_sales = vg_sales.loc[vg_sales.Platform == 'PS2'].groupby(['Publisher']).agg({'Global_Sales':'mean'}).sort_values('Global_Sales', ascending = False)
PS2_sales
```

```python
PS2_sales.head(20).plot.bar()
```

```python
pc_sales = vg_sales.loc[vg_sales.Platform == 'PC']
pc_sales.head(20)
```

```python
pc_sales.loc[vg_sales.Platform == 'PC'].groupby(['Name', 'Year', 'Genre']).agg({'Global_Sales':'sum'}).sort_values('Global_Sales', ascending = False)
```

```python
pc_sales_y = pc_sales.loc[vg_sales.Platform == 'PC'].groupby(['Name']).agg({'Global_Sales':'sum'}).sort_values('Global_Sales', ascending = False)
pc_sales_y.head(25).plot.bar()
```

```python
ws_sales = vg_sales.loc[vg_sales.Platform == 'WS'].groupby(['Name']).agg({'Global_Sales':'sum'}).sort_values('Global_Sales', ascending = False)

ws_sales_y = vg_sales.loc[vg_sales.Platform == 'WS'].groupby(['Name', 'Year']).agg({'Global_Sales':'sum'}).sort_values('Global_Sales', ascending = False)
ws_sales_y
```
